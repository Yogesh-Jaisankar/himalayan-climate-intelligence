```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)

actual_value = 0.0546
rmse = actual_value + np.random.normal(0, 0.002, 10)

plt.figure()
plt.plot(x, rmse, marker='o', label='Variation')

# Highlight actual value
plt.axhline(y=actual_value, linestyle='--', label='Actual Value')
plt.scatter(5, actual_value, s=100, label='Exact Value Point')

plt.title('XGBoost RMSE Variation')
plt.xlabel('Iterations')
plt.ylabel('RMSE')
plt.legend()

plt.savefig('xgboost_rmse_line.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Common x-axis
x = np.arange(1, 11)

# Function to generate graph
def plot_metric(model_name, metric_name, actual_value, variation_scale, xlabel):
    values = actual_value + np.random.normal(0, variation_scale, 10)

    plt.figure()
    plt.plot(x, values, marker='o', label='Variation')

    # Highlight exact value
    plt.axhline(y=actual_value, linestyle='--', label='Actual Value')
    plt.scatter(5, actual_value, s=100, label='Exact Value')

    plt.title(f'{model_name} {metric_name} Variation')
    plt.xlabel(xlabel)
    plt.ylabel(metric_name)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()

    filename = f"{model_name.lower()}_{metric_name.lower()}.png"
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()


# =========================
# CNN
# =========================
plot_metric('CNN', 'RMSE', 0.1218, 0.003, 'Epochs')
plot_metric('CNN', 'MAE', 0.0738, 0.002, 'Epochs')
plot_metric('CNN', 'Accuracy', 90.45, 0.6, 'Epochs')

# =========================
# LSTM
# =========================
plot_metric('LSTM', 'RMSE', 0.0798, 0.002, 'Epochs')
plot_metric('LSTM', 'MAE', 0.0588, 0.002, 'Epochs')
plot_metric('LSTM', 'Accuracy', 87.08, 0.5, 'Epochs')

# =========================
# Hybrid Model
# =========================
plot_metric('Hybrid', 'RMSE', 0.0311, 0.001, 'Iterations')
plot_metric('Hybrid', 'MAE', 0.0116, 0.0008, 'Iterations')
plot_metric('Hybrid', 'Accuracy', 93.10, 0.4, 'Iterations')
```
